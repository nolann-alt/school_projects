```python
# Importer les bibliothèques nécessaires
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Tracer un graphique de base

```python
# Exemple de base : une ligne simple
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Une ligne simple")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

## 2. Personnalisation d'un graphique

```python
# Ajout de styles et d'annotations
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, label="sin(x)", color="blue", linestyle="--")
plt.plot(x, y2, label="cos(x)", color="green", linestyle="-")
plt.title("Personnalisation d'un graphique")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.legend()
plt.grid(True)
plt.show()
```

## 3. Création de sous-graphiques

```python
# Travailler avec plusieurs graphiques
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# Premier graphique
axs[0].plot(x, y1, color="red")
axs[0].set_title("sin(x)")
axs[0].grid(True)

# Deuxième graphique
axs[1].plot(x, y2, color="blue")
axs[1].set_title("cos(x)")
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

## 6. Tracés de Suites
Cette section montre comment représenter des suites discrètes avec Matplotlib.

```python
def Syracuse(n):
 
    syrac=[]
 
    while n!=1:
        if n%2 == 0:
            n = n/2
            
        else:
            n = (n*3)+1
        syrac.append(n)   
    return syrac




syrac =  Syracuse(27)




plt.stem(np.arange(len(syrac)), syrac, basefmt=" ")
plt.title("Suite de Syracuse")
plt.xlabel("n")
plt.ylabel("a_n")
plt.grid(True)
plt.show()
```

```python
plt.plot(np.arange(len(syrac)), syrac, 'rx', markersize=4)
plt.title("Suite de Syracuse")
plt.xlabel("n")
plt.ylabel("a_n")
plt.grid(True)
plt.show()
```

## 4. Histogrammes et Scatter plots

```python
# Histogramme
data = np.random.randn(1000)

plt.hist(data, bins=30, color="purple", alpha=0.7)
plt.title("Histogramme")
plt.xlabel("Valeurs")
plt.ylabel("Fréquence")
plt.show()

# Scatter Plot
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, color="orange", alpha=0.5)
plt.title("Scatter Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## 5. Graphiques 3D

```python
from mpl_toolkits.mplot3d import Axes3D

# Création de données
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Tracé 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap="viridis")
ax.set_title("Graphique 3D")
plt.show()
```
